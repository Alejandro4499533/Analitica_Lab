# Actividad C. Limpieza y tratamiento de errores (25 puntos)

Crea `df_limpio` y realiza:
1. Limpieza de texto (`campus`, `modalidad`)
2. Normalización de etiquetas (ej. `slp centro`, `SLP Cntro`, `escolar`, `MIXTA`)
3. Eliminar duplicados por `id_estudiante`
4. Corregir rangos en `promedio` y `asistencia_pct` (0–100)
5. Corregir `creditos_aprobados <= creditos_inscritos`
6. Imputar faltantes (`edad`, `promedio`, `asistencia_pct`) con mediana

```python
df_limpio = df_crudo_eval.copy()

# 9.1 Quitar espacios
df_limpio["campus"] = df_limpio["campus"].astype(str).str.strip()
df_limpio["modalidad"] = df_limpio["modalidad"].astype(str).str.strip()

# 9.2 Normalizar y corregir etiquetas
df_limpio["campus"] = df_limpio["campus"].str.lower().map({
    "slp centro": "SLP Centro",
    "slp cntro": "SLP Centro",
    "slp norte": "SLP Norte",
    "matehuala": "Matehuala"
}).fillna(df_limpio["campus"])

df_limpio["modalidad"] = df_limpio["modalidad"].str.lower().map({
    "escolarizada": "Escolarizada",
    "escolar": "Escolarizada",
    "mixta": "Mixta"
}).fillna(df_limpio["modalidad"])

# 9.3 Eliminar duplicados por id
df_limpio = df_limpio.drop_duplicates(subset=["id_estudiante"], keep="first").copy()

# 9.4 Corregir rangos
df_limpio["promedio"] = df_limpio["promedio"].clip(0, 100)
df_limpio["asistencia_pct"] = df_limpio["asistencia_pct"].clip(0, 100)

# 9.5 Corregir error lógico
df_limpio["creditos_aprobados"] = np.minimum(df_limpio["creditos_aprobados"], df_limpio["creditos_inscritos"])

# 9.6 Imputación simple con mediana
for col in ["edad", "promedio", "asistencia_pct"]:
    df_limpio[col] = df_limpio[col].fillna(df_limpio[col].median())

print("Limpieza completada ✅")
display(df_limpio)
```

# Actividad E. Fusión (`merge`) y resumen final (15 puntos)

1. Une `df_limpio` con `apoyos` por `id_estudiante` con `how='left'` e `indicator=True`
2. Guarda en `df_final`
3. Muestra `shape` y conteo de `_merge`
4. Resume:
   - promedio general
   - promedio por `campus`
   - distribución de `riesgo_academico`
5. Guarda CSV en `salida_eval/estudiantes_preprocesados_eval.csv`

```python
# --- Respuesta del estudiante (E1) ---
df_final = pd.merge(df_limpio, apoyos, on="id_estudiante", how="left", indicator=True)
df_final
# Merge y creación de df_final
```

```python
# --- Respuesta del estudiante (E2) ---
print("Promedio general:", round(df_final["promedio"].mean(), 2))
print("Asistencia promedio:", round(df_final["asistencia_pct"].mean(), 2))

# Crear tasa_aprobacion en df_final
df_final["tasa_aprobacion"] = df_final["creditos_aprobados"] / df_final["creditos_inscritos"]

# Crear riesgo_academico en df_final
condiciones = [
    (df_final["promedio"] < 70) | (df_final["asistencia_pct"] < 75),
    (df_final["promedio"] < 80)
]
opciones = ["Alto", "Medio"]
df_final["riesgo_academico"] = np.select(condiciones, opciones, default="Bajo")

# Display the relevant columns from df_final including the newly created ones
display(df_final[["id_estudiante", "promedio", "asistencia_pct", "tasa_aprobacion", "riesgo_academico"]])

print("\nDistribución de riesgo:")
display(df_final["riesgo_academico"].value_counts())

print("\nPromedio por campus:")
display(df_final.groupby("campus")["promedio"].mean().round(2))
# Resumen final
```

```python
# --- Respuesta del estudiante (E3) ---
df_final.to_csv("salida_eval/estudiantes_preprocesados_eval.csv", index=False)
# Guardar archivo final en CSV
```
